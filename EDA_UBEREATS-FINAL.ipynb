### Filling in the missing Values

```python
#FILLING IN THE NULL VALUES OF SCORE

def fill_with_group_mean(df, group_cols, fill_cols):
    '''
    This function imputes the mssing values by grouping the rows by category and price range 
    and getting the mean score and filling respective group by their group means
    '''
    group_means = df.groupby(group_cols)[fill_cols].transform('mean')
    df[fill_cols] = df[fill_cols].fillna(group_means)
    # After filling with group mean, fill remaining NaNs with a default value
    df[fill_cols] = df[fill_cols].fillna(-1)  # Use a default value (-1 for No rating)
    return df

# Apply the function to impute only the score
res = fill_with_group_mean(res, ['category', 'price_range'], ['score'])

# Display info to verify
print(f"Row count after transform: {len(res)}")
print(res.info())
print(res.isnull().sum())
```

```python

# Function to set the score to -1 if the rating is -1
#-1 means no ratings were given to the restaurant, hence score is also -1
def correct_score_based_on_ratings(df):
    df.loc[df['ratings'] == -1, 'score'] = -1
    return df

# Apply the function to the DataFrame
res = correct_score_based_on_ratings(res)
```

```python
#Filling in the null values of ratings to -1
res['ratings'] = res['ratings'].fillna(-1) #-1 means no ratings were given
```

```python
#ALL the null values of object data type is imputed to Unknown 
res.fillna('Unknown', inplace = True)
# Fill missing descriptions with a placeholder
res_menu['description'] = res_menu['description'].fillna('No description available')
res_menu.dropna(subset=['name'], inplace=True)
```

```python

# Group by restaurant name and count unique full addresses
branch_counts = res.groupby('name')['full_address'].nunique()

# Filter restaurants with more than one unique address
multi_branch_restaurants = branch_counts[branch_counts > 1].index

# Filter restaurants with exactly one unique address
single_branch_restaurants = branch_counts[branch_counts == 1].index

# Filter the original DataFrame to show details of restaurants with a single branch
single_branch_details = res[res['name'].isin(single_branch_restaurants)]
# Filter the original DataFrame to show details of restaurants with multiple branches
multi_branch_details = res[res['name'].isin(multi_branch_restaurants)]

# Select only the required columns and explicitly create a new DataFrame
columns_to_select = ['id', 'name', 'full_address', 'lat', 'lng', 'zip_code', 'ratings', 'score','price_range','category']

single_branch_df = pd.DataFrame(single_branch_details, columns=columns_to_select)
single_branch_df.reset_index(drop=True, inplace=True)
multi_branch_df = pd.DataFrame(multi_branch_details, columns=columns_to_select)
multi_branch_df.reset_index(drop=True, inplace=True)
# Add 'branch' column with value 'Single' to single_branch_df
single_branch_df['branch'] = 'Single'

# Add 'branch' column with value 'Multiple' to multi_branch_df
multi_branch_df['branch'] = 'Multiple'
```

#### Finding out the top 20 best performing Multi-Branch Restaurants.

```python
#Grouping by the name and finding the top 20 performing multi-branch restaurants
grouped_restaurants = multi_branch_df.groupby('name').agg({
    'score': 'mean',
    'ratings': 'sum',
    'id': 'first',
    'price_range':'first',
    'full_address': 'first',
    'lat': 'first',
    'lng': 'first',
    'zip_code': 'first',
    'category':'first',
    'branch':'first'
}).reset_index()

# Step 2: Calculate the combined metric
grouped_restaurants['combined_metric'] = grouped_restaurants['score'] * grouped_restaurants['ratings']

# Step 3: Sort based on the combined metric
top_restaurants_multi = grouped_restaurants.sort_values(by='combined_metric', ascending=False).reset_index(drop=True)
print("Top 20 Best Performing Multi-Branch Restaurants")
# Step 4: Display top 10 restaurants based on combined metric
top_20_multi_branch_restaurants = top_restaurants_multi.head(20)
top_20_multi_branch_restaurants
```

```python
#To count the number of outlets for the multi-branch restaurants 
outlet_counts = multi_branch_df[multi_branch_df['name'].isin(top_restaurants_multi['name'])].groupby('name')['full_address'].nunique().reset_index()
outlet_counts.columns = ['name', 'no_of_outlets']
```

```python
# Merge the outlet counts with the top restaurants data
top_restaurants_multi = top_restaurants_multi.merge(outlet_counts, on='name').head(20)
```

#### Finding out the top 20 best performing Single-Branch Restaurants(Independent Restaurants).

```python
# Step 1: Filter out invalid scores and create a copy
valid_restaurants_single = single_branch_df[single_branch_df['score'] != -1].copy()


# Compute the combined metric
valid_restaurants_single['combined_metric'] = valid_restaurants_single['score'] * valid_restaurants_single['ratings']

# Step 3: Sort based on the combined metric
top_restaurants_combined_single = valid_restaurants_single.sort_values(by='combined_metric', ascending=False).reset_index(drop = True)

# Step 4: Display top 10 restaurants based on combined metric
print("Top 20 Best Performing Single-Branch Restaurants:")
top_restaurants_combined_single = top_restaurants_combined_single.head(20)

top_restaurants_combined_single
```
